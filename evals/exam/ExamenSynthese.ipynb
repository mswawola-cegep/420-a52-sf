# Partie 2 - Comprendre pourquoi les gens votent (50 points)

En août 2006, trois chercheurs (Alan Gerber et Donald Green de l'Université de Yale, et Christopher Larimer de l'Université du Nord de l'Iowa) ont mené une étude à grande échelle dans le Michigan (É.U.), pour tester l'hypothèse selon laquelle l'une des raisons pour lesquelles les gens votent est une pression sociale. Pour citer le premier paragraphe de leur document de recherche de 2008:

*Among the most striking features of a democratic political system is the participation of millions of voters in elections. Why do large numbers of people vote, despite the fact that ... "the casting of a single vote is of no significance where there is a multitude of electors"? One hypothesis is adherence to social norms. Voting is widely regarded as a citizen duty, and citizens worry that others will think less of them if they fail to participate in elections. Voters' sense of civic duty has long been a leading explanation of vote turnout...*

Les chercheurs ont échantillonné environ 344 000 électeurs dans différents groupes. Environ 191 000 électeurs constituaient un groupe "contrôle" et les autres ont été classés dans l'un des quatre groupes "de traitement". Ces cinq groupes correspondent à cinq variables binaires.

* "Civic Duty" (variable `civicduty`) personnes ayant reçu une lettre avec le message "DO YOUR CIVIC DUTY - VOTE!"

* "Hawthorne Effect" (variable `hawthorne`) personnes ayant reçu la même lettre que le groupe "Civic Duty") et le message supplémentaire "YOU ARE BEING STUDIED". Ces personnes ont également été informées que leur comportement électoral serait examiné au moyen de documents publics.

* "Self" (variable `self`) les membres du groupe ont aussi reçu le message « Civic Duty » ainsi que le relevé récent de vote de tous les membres de ce foyer et un message indiquant qu'un autre message serait envoyé après l'élection avec des relevés mis à jour.

* "Neighbors" (variable `neighbors`) personnes ayant reçu le même message que celui du groupe « Self », sauf que le message avait non seulement les dossiers de vote du foyer mais aussi celui des voisins - maximisant la pression sociale.

* "Control" (variable `control`) personnes n'ayant rien reçu, correspondant à une situation usuelle.

D'autres variables incluent le genre (`sex`, 0 homme, 1 femme), l'année de naissance (`yob`) et la variable dépendante `voting` (1 si vote, 0 si abstention).
